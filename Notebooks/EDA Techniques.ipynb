# Now we are going to handle Mising Values

```python
# in the column of "Age" there are two missing values
# we can simply add mean values there

import numpy as np

clean_data['Age'] = clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age'])))
```

```python
clean_data['Age']
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
```

```python
clean_data['Exp']
```

```python
clean_data['Location'] = clean_data['Location'].fillna(clean_data['Location'].mode()[0])
```

```python
clean_data['Location']
```

```python
clean_data
```

# This is how we fillied the Missing Values as well

```python
clean_data.info()
```

```python
clean_data['Age'] = clean_data['Age'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data['Exp'] = clean_data['Exp'].astype(int)
```

```python
clean_data['Salary'] = clean_data['Salary'].astype(str)                           # makes sure all values are treated as text.
clean_data['Salary'] = clean_data['Salary'].str.replace('O', '0')                 # fixes the typo-like letter O.
clean_data['Salary'] = clean_data['Salary'].str.replace(r'\D', '', regex=True)    # removes anything that is not a digit (%, ^, $, -, spaces, etc.).
clean_data['Salary'] = clean_data['Salary'].astype(int)                           # safely converts the cleaned strings to integers.
```

```python
clean_data.info()
```

```python
clean_data['Name'] = clean_data['Name'].astype('category')    
clean_data['Domain'] = clean_data['Domain'].astype('category')    
clean_data['Location'] = clean_data['Location'].astype('category')    
```

```python
clean_data.info()
```
